```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
housing_raw = pd.read_csv(
    'assignments/data/housing_data.csv',
    parse_dates=['period_begin', 'period_end'],
    dtype={'total_homes_sold': 'Int32', 'region_name': 'category'}
)
housing_raw.head()
```

```python
housing_raw.info()
```

```python
ca_housing = housing_raw.loc[
    (housing_raw['region_name'].str.contains('CA'))
].assign(region_name = housing_raw['region_name'].str[0:-11])
ca_housing.info(memory_usage='deep')
```

```python
ca_housing.head()
```

```python
ca_housing.shape
```

```python
ca_housing_pivot = ca_housing.pivot_table(
    index = 'period_begin',
    columns = 'region_name',
    values = 'median_active_list_price'
).assign(CA_average = lambda x: x.mean(axis=1))

ca_housing_pivot.head()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
fig.suptitle('CA Housing', fontsize=16)
ax.set_title('LA keeps rising')

columns = ['CA_average', 'San Francisco', 'Los Angeles']
for col in columns:
    ax.plot(ca_housing_pivot.index, ca_housing_pivot[col]/1000000, label=col)


ax.set_xlabel('Year')
ax.set_ylabel('Median home price (Millions)')
ax.set_ylim(0, 1.800000)

# ax.set_xticks(['2017-04-01', '2020-04-01', '2022-04-01'])
ax.set_xticks(ca_housing_pivot.index[::12])
plt.xticks(rotation=45)
ax.legend(loc='upper right')

ax.axvline(18341, c='black', ls='--')
ax.text(18520, 0.95, '... while prices elsewhere\nin California keep rising')

ax.annotate('Start of Covid',
           xy=(18330, 1.1),
           xytext=(17900, 1.1),
           arrowprops=dict(
               facecolor='black',
               width=.2,
               headwidth=5,
               connectionstyle='angle3, angleA=270, angleB=0'
           ),
           verticalalignment='center')

fig.text(0.3, 0.8, 'San Francisco prices hit hard bi Covid')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```
