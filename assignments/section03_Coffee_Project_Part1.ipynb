# Coffee Production Stats

* Create a bar chart of the top 10 coffee producers by total production in 'total-production.csv', sorted from highest to lowest. Change the xticks to two letter country abbreviations.


* Data Prep Code

<mark style="background-color: #ffffff"> coffee_production = pd.read_csv("coffee/total-production.csv").T </mark> 

<mark style="background-color: #ffffff"> coffee_production.columns = coffee_production.iloc[0] </mark> 

<mark style="background-color: #ffffff"> coffee_production = coffee_production.drop("total_production") </mark> 


```python
coffee_production = pd.read_csv(path).T

coffee_production.columns = coffee_production.iloc[0]
coffee_production = coffee_production.drop("total_production")

coffee_production.head()
```

```python
coffee_production.info()
```

```python
top10_producers = coffee_production.sum().sort_values(ascending=False).iloc[:10]

top10_producers
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

ax.bar(top10_producers.index,
      top10_producers / 1000000)

ax.set_ylim(0, 1.2)
ax.set_ylabel('Total Production (millions 60kg bags)')
ax.set_title('Top 10 production countries 1990-2018', fontsize=16)

ax.xaxis.set_tick_params(rotation=30)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```

# Top 5 over time

* Create a line chart with the top 5 nations over time. Make sure to include a legend! Consider converting the index to datetime64[ns].

```python
top5_countries = top10_producers.index[:5].tolist()
top5 = coffee_production[top5_countries]

top5.head()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

for c in top5_countries:
    ax.plot(top5.index,
           top5[c]/1000,
           label=c)

ax.set_ylim(0, 65)
ax.set_ylabel('Production (Thousand 60kg bags)')
ax.set_xticks(top5.index[2::4])

ax.set_title('Vietnam surges to Number 2 Spot')
fig.suptitle('Top coffe producing Nations 1990-2018')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.legend()
plt.show()
```

# COMPOSITION OVER TIME

Our client is interested in looking at how Brazil's share of global coffee production has changed over time.

Build a plot that global total production as well as Brazil's contribution to global production over time.

```python
brazil_vs_others = coffee_production.assign(
    rest_of_world = coffee_production.drop('Brazil', axis=1).sum(axis=1)
).loc[:, ['Brazil', 'rest_of_world']].astype({'Brazil': 'float64'})

brazil_vs_others
```

```python
brazil_vs_others.head()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

ax.stackplot(brazil_vs_others.index,
            brazil_vs_others['Brazil'],
             brazil_vs_others['rest_of_world']
            )

ax.set_ylim(0, 180000)
ax.set_ylabel('Production (60kg bags)')
ax.legend(['Brazil', 'Others'], loc='upper left')
ax.set_xticks(brazil_vs_others.index[2::4])

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```

# Coffee Consumption Vs. Prices

Compare coffee prices with coffee consumption in the following countries. 

Is there a relationship between coffee consumption and the price these countries pay?

Plot a dual axis bar chart of price vs consumption totals for these countries.

```python
consumption = pd.read_csv("./data/coffee/imports.csv").set_index('imports').mean(axis=1)

consumption.name = 'imports'

consumption.head()
```

```python
prices = pd.read_csv("./data/coffee/retail-prices.csv").set_index('retail_prices').mean(axis=1)

prices.name = 'prices'

prices.head()
```

```python
price_cons = (pd.DataFrame(prices)
              .merge(consumption, left_on=prices.index, right_on=consumption.index, how='inner')
             ).sort_values('imports', ascending=False)

price_cons.columns = ['country', 'Price', 'Imports']
price_cons['country'] = price_cons['country'].str.strip(' ')
price_cons
```

```python
price_cons
```

```python
import numpy as np

fig, ax = plt.subplots(figsize=(15, 10))

x = np.arange(price_cons.shape[0])
width = .25

bar1 = ax.bar(x-width/2,
       price_cons['Imports'],
       label='Imports',
       width=width
      )

ax2 = ax.twinx()
bar2 = ax2.bar(x+width/2,
       price_cons['Price'],
       label='Price',
       width=width,
        color='orange'
      )

ax.set_xticks(x)
ax.set_xticklabels(price_cons.country)

ax.legend([bar1, bar2], ['Consumption', 'Price'])

plt.show()
```
