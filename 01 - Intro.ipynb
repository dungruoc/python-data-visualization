```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = [x * 10 for x in range(10)]
data
```

```python
x_labels = [f'{x}!' for x in range(10)]
x_labels
```

```python
plt.plot(x_labels, data)
plt.show()
```

```python
df = pd.DataFrame(data, x_labels).assign(series2 = lambda x: x[0] / 2)
df
```

```python
plt.plot(df)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(df)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(df)
fig.suptitle('Fig title')
ax.set_title('ax title')
plt.show()
```
